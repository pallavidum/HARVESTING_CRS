```python
def plot_distribution(result):
    std_left = (np.median(result) - np.std(result))
    std_right = (np.median(result) + np.std(result))
    print("max:", np.max(result))
    print("min:", np.min(result))
    print("mean:", np.mean(result))
    print("median:", np.median(result))
    print("std:", np.std(result))
    print(std_left)
    print(std_right)
    plt.hist(result, bins=55)
    plt.axvline(x=np.mean(result), color='cyan')
    plt.axvline(x=np.median(result), color='red')
    plt.axvline(x=std_left, color='orange')
    plt.axvline(x=std_right, color='green')
    
def plot_get_bins(result_removed_outliers):
    std_left = np.median(result_removed_outliers) - np.std(result_removed_outliers)
    std_right = np.std(result_removed_outliers) + np.median(result_removed_outliers)
    (n, bins, patches) = plt.hist(result_removed_outliers, bins=55)
    #plt.axvline(x=np.mean(result_removed_outliers), color='cyan')
    plt.axvline(x=np.median(result_removed_outliers), color='red')
    plt.axvline(x=std_left, color='orange')
    plt.axvline(x=std_right, color='green')
    return bins
    
def compute_score(num):
    for index, elem in enumerate(bins):
        if elem > num:
            return (550- index * 10) + 300
    return 300

def compute_score_standard(proba):
    factor = 20/np.log(2)
    intercept = 600 - factor * np.log(50)
    score = np.round(np.log((1 - proba) / proba) * factor + intercept) # * 28.85 + 487.122
    if score < 0:
        print(score)
        return 0
    else:
        return score
    
plot_distribution(proba)
```

```python
std_right = (np.median(proba) + np.std(proba))
result_removed_outliers = [i for i in proba if i < std_right]
print(np.max(result_removed_outliers))
bins = plot_get_bins(result_removed_outliers)
```

```python
df['score'] = df.apply(lambda x: compute_score(x['default_proba']), axis=1)
df['score_standard'] = df.apply(lambda x: compute_score_standard(x['default_proba']), axis=1)
df
```
