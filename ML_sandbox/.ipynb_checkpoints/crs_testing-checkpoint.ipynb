```python
def coarse_classifier(df, col, ranges=[]):
    if not ranges:
        return "ranges cannot be empty"
    df['bin'] = pd.cut(df[col], ranges)
    return df['bin'].values
```

```python
# combines pay features into single column
def combine_pay_feature(df):
    cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
    return np.round((df[cols[0]] + df[cols[1]] + df[cols[2]] + df[cols[3]] + df[cols[4]] + df[cols[5]]) / 6)

df['BILL_COMBINED'] = df.apply(combine_pay_feature, axis=1)
df.drop(['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6'], axis=1, inplace=True)
df.head()
```

```python
# testing gradient boosting with undersampling/oversampling
# TODO try tuning hyperparams

CR = CreditRisk("", "")
df, classes = CR.read_file(df, "")
X, y = CR.test_train_matrix(df)

#X_resampled, y_resampled = RandomOverSampler().fit_sample(X, y) # = 69 +/- 7
#X_resampled, y_resampled = SMOTE().fit_sample(X, y) # = 86 +/- 7
#X_resampled, y_resampled = ADASYN().fit_sample(X, y) # = 86 +/- 7

#X_resampled, y_resampled = RandomUnderSampler().fit_sample(X, y) # = 64 +/- 8
#X_resampled, y_resampled = EditedNearestNeighbours().fit_sample(X, y) # = 75 +/- 16

#X_resampled, y_resampled = SMOTEENN().fit_sample(X, y) # = 86 +/- 15
X_resampled, y_resampled = SMOTETomek().fit_sample(X, y) # = 87 +/- 7

print(len(y_resampled))
counts = Counter(y_resampled)
print(counts)

features = list(df)
#print(features)
X_train, X_test, y_train, y_test = CR.split_train_test(
            X_resampled, y_resampled, 0.20, False, len(features))

model = XGBClassifier(nthread=-1) # -1 to use all cores in system
model.fit(X_train, y_train)
y_train_pred = model.predict_proba(X_train)[:, 1] # TODO 1 = yes default, 0 = no default?
y_test_pred = model.predict_proba(X_test)[:, 1]

y_pred = model.predict(X_test)

print("gradient boosted trees accuracy: " + str(CR.get_model_accuracy(model, X_test, y_test)))

kfold = KFold(n_splits=10)
results = cross_val_score(model, X_resampled, y_resampled, cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))

kfold = StratifiedKFold(n_splits=10)
results = cross_val_score(model, X_resampled, y_resampled, cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))

print("ROC AUC Score:", roc_auc_score(y_test, y_test_pred))
fpr, tpr, threshold = roc_curve(y_test, y_test_pred)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

print("F1 Score:", f1_score(y_test, y_pred, average='binary'))

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("tn:", tn, "fp:", fp, "fn:", fn, "tp:", tp)
confusion_matrix(y_test, y_pred)

log_loss("log loss:", y_test, y_test_pred)
```
